## Edge filter

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Sobel kernel with numpy
kernel = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                      [-1, 0, 1]])
# Read image
img = cv2.imread('cup.jpeg')
plt.imshow(img)
plt.show()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply kernel
sobelX = cv2.filter2D(gray, -1, kernel)

# Show image
plt.imshow(sobelX, cmap='gray')
plt.show()

# Sobel kernel in Y direction
kernel = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply kernel
sobelY = cv2.filter2D(gray, -1, kernel)

# Show image
plt.imshow(sobelY, cmap='gray')
plt.show()

#Magnitude of the gradient
sobel = np.sqrt(sobelX**2 + sobelY**2) #this is the square root of(x**2 +y**2)
for i in range(sobel.shape[0]):
    for j in range(sobel.shape[1]):
        sobel[i,j]=(sobelX[i,j]**2+sobelY[i,j]**2)**0.5 #the power 0.5 means square root

# Show image
plt.imshow(sobel, cmap='gray')
plt.show()


```

## Sobel filter

```python
#Sobel filter 

#read image
img = cv2.imread('cup.jpeg')
img=cv2.GaussianBlur(img,(3,3),100)
plt.imshow(img)
plt.show()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply one shot filter
sobel = cv2.Sobel(gray, cv2.CV_64F,1, 1, ksize=5)

#normalize sobel to range 0 to 255
sobel = abs(sobel)

# Show image
plt.imshow(sobel, cmap='gray')
plt.show()

```

## Line detection

```python
# apply hough transform to image shapes.jpg
img = cv2.imread('shapes.jpg')
img=cv2.GaussianBlur(img,(3,3),100)
plt.imshow(img)
plt.show()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply one shot filter
canny = cv2.Canny(gray, 2, 80)

# Show image
plt.imshow(canny, cmap='gray')
plt.show()

# Apply hough transform
lines = cv2.HoughLines(canny, 1, np.pi/180, 50) # the first param should be a binary image while the fourth param represents the minimum number of votes
                                                # which means the minimum number of points on the line which means the minimum length of a line
                                                # the second param represents rooo while third represents theta

# Draw lines
for line in lines:
    print(line)
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    
    #draw line
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    
    #draw line
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    
    #draw line
    cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Show image
plt.imshow(img) # the lines will be blue because the plt shows images by RGB model not BGR
plt.show()
```

```python
# apply hough transform to image shapes.jpg
img = cv2.imread('shapes.jpg')
img=cv2.GaussianBlur(img,(3,3),100)
plt.imshow(img)
plt.show()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Apply one shot filter
canny = cv2.Canny(gray, 2, 80)

# Show image
plt.imshow(canny, cmap='gray')
plt.show()
# Apply hough transform
lines = cv2.HoughLinesP(canny, 1, np.pi/180, 10)

# Draw lines
for line in lines:
    print(line)
    x0, y0, x1, y1 = line[0]
    cv2.line(img, (x0, y0), (x1, y1), (0, 250, 250), 3)
    
# Show image
plt.imshow(img)
plt.show()
```

## Hough Circle Detection

```python
# apply circle hough transform to image shapes.jpg
img = cv2.imread('shapes.jpg')
img=cv2.GaussianBlur(img,(3,3),100)
plt.imshow(img)
plt.show()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply one shot filter
canny = gray

# Apply hough transform
circles = cv2.HoughCircles(canny, cv2.HOUGH_GRADIENT,1, 20, param1=100, param2=45, minRadius=1, maxRadius=200)

# Draw circles
for circle in circles[0, :]:
    print(circle)
    x, y, r = circle
    #convert to int
    x = int(x)
    y = int(y)
    r = int(r)

    cv2.circle(img, (x, y), r, (0, 255, 0), 4)
# Show image
plt.imshow(img)
plt.show()
```

# Contours Detection

```python

# Create Mask of contours
img = cv2.imread('shapes.jpg')
img=cv2.GaussianBlur(img,(3,3),100)
plt.imshow(img)
plt.show()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply one shot filter
canny = cv2.Canny(gray, 2, 80)

# Show image
plt.imshow(canny, cmap='gray')
plt.show()

# Apply detection
contours, hierarchy = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Create masks
for contour in contours:
    # Create empty mask
    mask = np.zeros(img.shape[:2], np.uint8)
    # Draw contour on mask
    cv2.drawContours(mask, [contour],-1, 255, -1)#the mask is the image that I am gonna draw the contour on while the [contour] 
                                                 #Indicates the contours obtained from the findContours() function and the -1 to say that I am gonna
                                                 #draw all contour points /255 is white colour/-1 is supposed to be the thickness  but here it fill the shape
    
    # Show mask
    plt.imshow(mask, cmap='gray')
    plt.show()
    
    # Apply mask to image
    masked = cv2.bitwise_and(img, img, mask=mask)
    # Show image
    plt.imshow(masked)
    plt.show()

```

```python
plates=cv2.imread('Screenshot 2023-09-03 143359.png')
plates=cv2.GaussianBlur(plates,(3,3),100)
plt.imshow(plates)
plt.show()

def img_show(img):
    cv2.imshow('Plates',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python

img_show(plates)

platesGray = cv2.cvtColor(plates, cv2.COLOR_BGR2GRAY)
img_show(platesGray)

canny_plates= cv2.Canny(platesGray, 10,500)
img_show(canny_plates)

#contours, hierarchy = cv2.findContours(canny_plates,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
contours, hierarchy = cv2.findContours(canny_plates, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)



for contour in contours:
    zeros = np.zeros(plates.shape)
    cv2.drawContours(zeros, [contour], -1, [255,255,255], -1)
    img_show(zeros)
    
```

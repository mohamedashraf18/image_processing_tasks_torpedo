```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# apply hough transform to image shapes.jpg
img = cv2.imread('Task 1/fourth.png')
img=cv2.GaussianBlur(img,(3,3),100)

img2=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply one shot filter(CANNY EDGE DETECTION)
canny = cv2.Canny(gray, 10, 50) #some how it did not recognize the main bottom line below when I was making the threshold (10,80)

# Show image
plt.imshow(canny, cmap='gray')
plt.show()

# Apply hough transform
lines = cv2.HoughLines(canny, 1, np.pi/180, 50) # the first param should be a binary image while the fourth param represents the minimum number of votes
                                                # which means the minimum number of points on the line which means the minimum length of a line
                                                # the second param represents rooo while third represents theta

# Draw lines
for line in lines:
    print(line)
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    
    #draw line
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    
    #draw line
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    
    #draw line
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 255), 2)

# Show image
plt.imshow(img) # the lines will be blue because the plt shows images by RGB model not BGR
plt.show()
```
